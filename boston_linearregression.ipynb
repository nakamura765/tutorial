```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston_df = DataFrame(boston.data)
boston_df.head()
```

```python
print(boston.feature_names)
```

```python
boston_df.columns = boston.feature_names

boston_df.head()
```

```python
feature_names_japanese = ['犯罪', '住宅密度', '商業割合', '川周辺', '窒素酸化物', '部屋数', '古物件', '雇用施設距離', '高速アクセス', '不動産税率', '教師率', '黒人率', '低給職業']
boston_df.columns = feature_names_japanese

boston_df.head()
```

```python
boston_df.shape
```

```python
boston_df['価格'] = DataFrame(boston.target)
boston_df.head()
```

```python
df = boston_df
df.info()
```

```python
#　本当はここでseabornでheatmapを出したいですが日本語の文字化け対処ができていないのでとばしました
```

```python
df.corr()['価格']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = df['価格']
X = df.drop('価格', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print('決定係数(train):{:.4f}'.format(model.score(X_train,y_train)))
print('決定係数(test):{:.4f}'.format(model.score(X_test,y_test)))
```

```python
print('\n回帰係数\n{}'.format(pd.Series(model.coef_, index=X.columns)))
print('\n切片: {:3f}'.format(model.intercept_))
```
