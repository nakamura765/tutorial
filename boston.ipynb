```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston_df = DataFrame(boston.data)
boston_df.head()
```

```python
print(boston.feature_names)
```

```python
boston_df.columns = boston.feature_names

boston_df.head()
```

```python
feature_names_japanese = ['犯罪', '住宅密度', '商業割合', '川周辺', '窒素酸化物', '部屋数', '古物件', '雇用施設距離', '高速アクセス', '不動産税率', '教師率', '黒人率', '低給職業']
boston_df.columns = feature_names_japanese

boston_df.head()
```

```python
boston_df.shape
```

```python
boston_df['価格'] = DataFrame(boston.target)
boston_df.head()
```

```python
df = boston_df
df.info()
```

```python
df.corr()['価格']
```

```python
from sklearn import linear_model
```

```python
y = df['価格']
x = df.drop(['価格'], axis = 1)
```

```python
clf = linear_model.LinearRegression()

clf.fit(x,y)
```
